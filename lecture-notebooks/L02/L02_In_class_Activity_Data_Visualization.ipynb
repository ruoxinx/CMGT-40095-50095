### **Data Types**

In Python, data types are the classifications we give to different kinds of data elements. Here are a few basic types:

- **Integers**: Whole numbers such as 1, 100, -50.
- **Floats**: Decimal numbers like 3.14, -0.001.
- **Strings**: Collections of characters, e.g., 'Hello', 'Data'.
- **Lists**: Ordered collections of items, e.g., [1, 2, 3] or ['a', 'b', 'c'].
- **Dictionaries**: Key-value pairs, e.g., {'name': 'John', 'age': 30}.

## **Sample Data Loading**
We'll use a fictional dataset representing construction costs and duration for projects across different states.

```python
# Creating a sample dataset
data = {
    'State': ['California', 'Texas', 'Florida', 'New York', 'Pennsylvania'],
    'Construction_Cost': [180000, 150000, 130000, 165000, 160000],
    'Project_Duration': [24, 18, 15, 22, 20]  # Duration in months
}

# Converting dictionary to DataFrame
df = pd.DataFrame(data)
```

## **Loading Data with Pandas**

Data is crucial in construction for making informed decisions. pandas makes it easy to load and manipulate data. Here's how to load data from a CSV file.

```python
data = pd.read_csv('construction_data.csv')
print("Data loaded successfully!")
print(data.head())  # Print the first five rows of the data
```

Here's how to load data from an Excel file.

```python
import pandas as pd

# Load the Excel file
excel_path = 'construction_data.xlsx'
data = pd.read_excel(excel_path)

# Display the first few rows of the DataFrame
print(data.head())
```

## **Basic Data Visualization**
Visualizing data helps in understanding complex patterns. Let’s create a simple bar chart.
[link text](https://)
### **Bar Chart: Comparing Construction Costs**
Bar charts are great for comparing quantities among different groups.

```python
plt.figure(figsize=(10, 6))
plt.bar(df['State'], df['Construction_Cost'], color='teal')
plt.xlabel('State')
plt.ylabel('Construction Cost ($)')
plt.title('Comparative Construction Costs by State')
plt.show()
```

## **Intermediate Visualization**
Let’s dig a bit deeper and combine our data points in a scatter plot to see if there’s a relationship between cost and duration.

### **Scatter Plot: Cost vs. Duration**

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Construction_Cost', y='Project_Duration', hue='State', data=df, s=100)
plt.title('Cost vs. Duration Across States')
plt.xlabel('Construction Cost ($)')
plt.ylabel('Project Duration (months)')
plt.legend(title='State')
plt.show()
```
